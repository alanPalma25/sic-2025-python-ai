1.1 Carga un conjunto de datos utilizando una biblioteca de Python como Pandas.

```python
# Espacio para desarrollar el código necesario.

# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Importar el data set
data = pd.read_csv('weatherAUS.csv')

# Mostrar las primeras filas del data set
data.head()
```

```python
# Estadistticas clave
data.describe()
```

```python
# Correlacion de las variables (drop no númericas)
data.select_dtypes(include=[np.number]).corr()
```

```python
# Distribucion de algunas variables (subplots)

data.select_dtypes(include=[np.number]).hist(bins=15, figsize=(15, 10))
plt.tight_layout()
plt.show()
```
